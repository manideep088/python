# Adding elements to dictionary

```python
#empty dictionaty
dict1={}
print(dict1)
```

Adding elements one at a time

```python
dict1[0]='Geeks'
dict1[1]='For'
dict1[2]='geeks'
dict1[3]=10
print(dict1)
```

Adding set of values to a single key

```python
dict1['value_set']=1,2,3,4,
print(dict1)
```

Updating a existing value

```python
dict1[0]='welcome'
dict1[1]='to'
print(dict1)
```

Adding Nested Key to Dictionaries

```python
dict1[5]={'Nested':{'1':'Life','2':'Python'}}
print(dict1)
```

# accessing elements of a Dictionary
In order to access the items of a dictionary refer to its key name

creating a Dictionary

```python
Dict1={1:'geeks',2:'for',3:'Geeks'}
print(Dict1)
```

```python
print(Dict1[1])
print(Dict1[3])
```

There is also a method called get() that will also help in accessing the element from a dictionary.This method accepts key as argument and returns the value.

```python
print(Dict1.get(3))
```

# Accessing a element of a Nested dictionary

```python
Dict={'dict1':{1:'Geeks'},
     'dict2':{'name':'manideep'}}
print(Dict['dict1'])
print(Dict['dict1'][1])
print(Dict['dict2']['name'])
```
